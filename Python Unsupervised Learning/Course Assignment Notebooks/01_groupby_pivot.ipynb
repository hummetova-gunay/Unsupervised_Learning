```python
import pandas as pd
```

```python

songs_dict = {
                'Customer': ['Aria', 'Aria', 'Aria', 'Chord', 'Chord', 'Harmony', 'Harmony', 'Harmony', 'Melody', 'Reed'],
                'Genre': ['Pop', 'Indie', 'Rock', 'Pop', 'Indie', 'Pop', 'Indie', 'Rock', 'Rock', 'Rock'],
                '# Songs': [50, 48, 1, 15, 36, 10, 5, 3, 2, 5]
             }

df = pd.DataFrame(songs_dict)
df
```

```python
# groupby

df.groupby('Customer')['# Songs'].sum().reset_index()
```

```python
# save as a dataframe

songs_groupby = df.groupby("Customer")['# Songs'].sum().reset_index()
songs_groupby
```

```python
# pivot

songs_pivot = df.pivot(index = 'Customer', columns ='Genre', values = '# Songs').fillna(0).reset_index()
songs_pivot
```
