#### handling missing values

```python
# dropna

customers.dropna().reset_index(drop=True)
```

```python
# save 

customers_dropped = customers.dropna().reset_index(drop = True)
customers_dropped.isna().any()
```

```python
# fillna 

customers.isna().any()
```

```python
# fill age

print(f'median age: {customers.Age.median()}')

customers['Age']=customers['Age'].fillna(round(customers['Age'].median()))

customers['Age']
```

```python
# fill followers (check with domain experts)
print(f'followers: {customers.Followers}')

customers['Followers'] = customers['Followers'].fillna(0)
customers['Followers']
```
