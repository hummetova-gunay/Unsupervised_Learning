```python
nutrition= pd.read_csv('../Data/fruit_nutrition.csv')
nutrition = nutrition.set_index('fruit_100g')
```

```python
nutrition.index.name = None
```

```python
nutrition
```

```python
cos_matrix_all = cosine_similarity(nutrition)
cos_df_all = pd.DataFrame(cos_matrix_all, index = nutrition.index, columns = nutrition.index)

cos_df_all
```

```python
# wrap in a function
def recommend_fruits(fruit_name, similarity_df):
    if fruit_name not in similarity_df.index:
        return f'{fruit_name} is not found in the fruit list'
    return similarity_df[[fruit_name]].sort_values(by = fruit_name, ascending = False)
 
```

```python
recommend_fruits('Mango', cos_df_all)
```

```python
recommend_fruits('Apple', cos_df_all)
```
